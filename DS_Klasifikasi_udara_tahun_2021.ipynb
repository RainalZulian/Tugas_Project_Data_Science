```python
import pandas as pd #library untuk dataframe
import numpy as np #library untuk komputasi
```

```python
# Import data ISPU 2021 hasil cleaning

df = pd.read_csv('ISPU_DKI_Jakarta_2021.csv')
df
```

```python
# Cek missing value

df.isnull().sum()
```

```python
# Cek kelas data kategori

df.categori.unique()
```

```python
df_2021 = df.copy()
```

```python
# Replace string to NaN

for kolom in df_2021.drop(columns=['tanggal','stasiun','critical']).columns:
  df_2021[kolom] = df_2021[kolom].replace(r'---',np.NaN,regex=True)

df_2021
```

```python
df_2021.info()
```

```python
df_2021.isna().sum()
```

```python
# Drop data mengandung NaN

df_2021 = df_2021.dropna(axis=0)
df_2021.isnull().sum()
```

```python
# Konversi tipe data polutan str menjadi int

df_2021 = df_2021.astype({'pm10': int, 'pm25': int, 'so2': int, 'co': int, 'o3': int, 'no2': int, 'max': int})

df_2021.info()
```

```python
from pyspark.sql import SparkSession
from pyspark.ml.feature import StringIndexer

# Inisialisasi sesi Spark
spark = SparkSession.builder.master("local[1]").appName("LabelEncodingExample").getOrCreate()

# Membuat DataFrame Spark dari pandas DataFrame
df_spark = spark.createDataFrame(df)

# Label Encoding menggunakan StringIndexer
string_indexer = StringIndexer(inputCol="categori", outputCol="categori_encoded")
model = string_indexer.fit(df_spark)
df_encoded = model.transform(df_spark)

# Tampilkan DataFrame hasil encoding
df_encoded.show()
```
